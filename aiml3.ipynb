Linear regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
t=np.array([5,7,12,16,20]).reshape(-1,1)
m=np.array([40,120,180,210,240])
```

```python
LR = LinearRegression()
LR.fit(t,m)
```

```python
time=float(input("enter time to pridict"))
out=LR.predict([[time]])
print("the mass is",out[0])
```

```python

plt.scatter(t,m,marker=".",label="Actual data points")
modelpred=LR.predict(t)
print(modelpred)
plt.plot(t,modelpred,color="black",label="RegressionLine")
plt.legend()
plt.title("Time and Mass")
plt.xlabel("Time")
plt.ylabel("Mass")
```
